```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load the CSV files
vegetation_df = pd.read_csv('dbg_vegetation_copy.csv')
npp_df = pd.read_csv('dbg_npp_copy.csv')
```

```python
# Calculate descriptive statistics for vegetation data
vegetation_stats = vegetation_df.describe()
print("Vegetation Data Descriptive Statistics:")
print(vegetation_stats)

# Calculate descriptive statistics for NPP data
npp_stats = npp_df.describe()
print("\nNPP Data Descriptive Statistics:")
print(npp_stats)

# Create a figure for Vegetation Data
fig, axes = plt.subplots(2, 3, figsize=(18, 10))
axes = axes.flatten()

for i, col in enumerate(vegetation_df.columns[1:]):  # Skip 'system:index'
    axes[i].boxplot(vegetation_df[col])
    axes[i].set_title(f'Boxplot of {col}')
    axes[i].set_ylabel('Percent pixel cover')

# Remove any empty subplots
for j in range(i+1, len(axes)):
    fig.delaxes(axes[j])

fig.tight_layout()
plt.suptitle('Vegetation Data Boxplots', y=1.02)
plt.show()

# Create a figure for NPP Data
fig, axes = plt.subplots(2, 2, figsize=(12, 10))
axes = axes.flatten()

for i, col in enumerate(npp_df.columns[1:]):  # Skip 'system:index'
    axes[i].boxplot(npp_df[col])
    axes[i].set_title(f'Boxplot of {col}')
    axes[i].set_ylabel('NPP (g C)')

# Remove any empty subplots
for j in range(i+1, len(axes)):
    fig.delaxes(axes[j])

fig.tight_layout()
plt.suptitle('NPP Data Boxplots', y=1.02)
plt.show()
```
