```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
labels = 'Gap', 'Canopy'
sizes = [156.15, 23.85]

fig, ax = plt.subplots(figsize=(10,10))
ax.pie(sizes, labels=labels, autopct='%1.1f%%', colors=['c', 'darkslategray'])
```

```python
labels = 'Shrub', 'Grass', 'Dead'
sizes = [21, 1, 3]

fig, ax = plt.subplots(figsize=(10,10))
ax.pie(sizes, labels=labels, autopct='%1.1f%%', colors = ['goldenrod', 'greenyellow', 'sienna'])
```

```python
labels = 'GPI', 'LPI'
sizes = [13.25, 13.88889]

fig, ax = plt.subplots(figsize=(8,6))
ax.bar(labels, sizes, color = ['lime', 'fuchsia'])
ax.set_ylabel('percent')
```
